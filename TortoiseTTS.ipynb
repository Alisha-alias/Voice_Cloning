```python
# This is the text that will be spoken.
text = "Let's strive to make the world a better place, one code block at a time."

# Pick a "preset mode" to determine quality. Options: {"ultra_fast", "fast" (default), "standard", "high_quality"}.
preset = "high_quality"

# Define the custom voice name
CUSTOM_VOICE_NAME = "custom"

# Create the folder for the custom voice
custom_voice_folder = f"tortoise/voices/{CUSTOM_VOICE_NAME}"
os.makedirs(custom_voice_folder, exist_ok=True)

# Define the expected file name
expected_file_name = "The Man in the arena.wav"

# Upload the file
uploaded_file = files.upload()

# Handle the uploaded file
if expected_file_name in uploaded_file:
    file_data = uploaded_file[expected_file_name]
    with open(os.path.join(custom_voice_folder, expected_file_name), 'wb') as f:
        f.write(file_data)
        print(f'Uploaded file "{expected_file_name}" to "{custom_voice_folder}"')
else:
    print(f'Error: The expected file "{expected_file_name}" was not uploaded.')
```

```python
voice_samples, conditioning_latents = load_voice(CUSTOM_VOICE_NAME)
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                          preset=preset)
torchaudio.save(f'generated-{CUSTOM_VOICE_NAME}.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio(f'generated-{CUSTOM_VOICE_NAME}.wav')
```
